## For Statement

```python
I=(1,2,3)
```

```python
for e in I:
    print(e)
```

```python
for i in [1,2,3]:
    print(i)
```

```python
for i in range(2,10):
    print(i,end=" ")#加空格的意思
```

```python
for x in range(2,10,3):
    print(x)
    
#(從2開始,小於10,每次加3)
```

```python
for i in range(10,3,-1):
    print(i,end=" ")
```

```python
c=0
a="abcdefghijklmnopqrstuvwxyz"
for i in a:
    c+=1
    print(i,end=" ")
    if c%10==0: #每10個字做一次
        print() #斷行
```

## While Statement

```python
counter = 6
while counter < 10:
    print(counter)
    counter += 1
```

```python
sum=0
i=1
while i<=100:
    sum+=i
    i+=1
print("1+2+3+......+99+100=",sum)
```
