```python
from skimage import io,data 
img=data.coffee() 
rows,cols,dims=img.shape 
img2=img[:,:,0].copy()#降維 0-R,1-G,2-B

threshold = 128
for i in range(0,rows):
    for j in range(0,cols):
        if img2[i,j] > threshold:
            img2[i,j] = 255 
        else:
            img2[i,j] = 0
print(img2.shape)
io.imshow(img2)
io.show()
```

```python
from skimage import data
import matplotlib.pyplot as plt 
import numpy as np

img=data.coffee()
rows,cols,dims=img.shape

threshold = 128

img2 = np.uint8(np.zeros((rows,cols,3)))
plt.figure(figsize=(8,8))
plt.subplot(121)
plt.title('origin image')
plt.imshow(img)

for i in range(0,rows): 
    for j in range(0,cols):
        if img[i,j,0] > threshold: 
            for k in range(0,3):
                img2[i,j,k] = 255
            else:
                for k in range(0,3): 
                    img2[i,j,k] = 255
print(img2.shape)
plt.subplot(122)
plt.title('grey image')
plt.imshow(img2)
plt.show()
```

```python
from skimage import io,data 
import matplotlib.pyplot as plt 
import numpy as np 
img=data.coffee() 
io.imshow(img)

rows,cols,dims=img.shape
half = np.matrix([[0,128],[192,64]]) #2*2 dithering 
img2=img[:,:,0].copy()

for i in range(0,rows): 
    for j in range(0,cols):
        index_i = i%2
        index_j = j%2
        if img2[i,j] > half[index_i, index_j]:
            img2[i,j] = 255 
        else:
            img2[i,j] = 0
plt.figure()
io.imshow(img2)
io.show()
```
