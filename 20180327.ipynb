```python
from skimage import io,data
import matplotlib.pyplot as plt
import numpy as np

img=data.coffee()
img2=img.copy()
img3=img.copy()
img4=img.copy()

plt.figure(num='haha',figsize=(20,20))

plt.subplot(141)
plt.title('origin image')
plt.imshow(img)

# modified image (RGB~0120red,green,blue)

plt.subplot(142)
plt.title('blue-green')
for i in range(50,400):
    for j in range(100,500):
        img2[i,j,0]=0
plt.imshow(img2)

plt.subplot(143)
plt.title('purple')
for i in range(50,400):
    for j in range(100,500):
        img3[i,j,1]=0
plt.imshow(img3)

plt.subplot(144)
plt.title('yellow')
for i in range(50,400):
    for j in range(100,500):
        img4[i,j,2]=0
plt.imshow(img4)

plt.show()
```
